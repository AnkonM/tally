```python
#5
year = int(input("Year: "))

def divisible(x,y):
    if not x%y:
        return True
    else:
        return False

def leap(year):
    if divisible(year, 4):
        if divisible(year, 100):
            if divisible(year, 400):
                return True
            else:
                return False
        else:
            return True
    else:
        return False

if leap(year):
    print("{0} is a leap year.".format(year))
else:
    print("{0} is not a leap year.".format(year))
```

```python
#7
num = int(input("Enter a two digit number: "))
if len(str(num))==2:
    pass
else:
    while True:
        num = int(input("Number needs to be two digit. Enter a TWO digit number: "))
        if len(str(num)) == 2:
            break 


numreverse = int(str(num)[-1] + str(num)[0])

print(numreverse)
```

```python
#8
num = int(input("Enter a three digit number: "))
if len(str(num))==3:
    pass
else:
    while True:
        num = int(input("Number needs to be three digit. Enter a THREE digit number: "))
        if len(str(num)) == 3:
            break 


numreverse = int(str(num)[-1] + str(num)[1] + str(num)[0])

print(numreverse)
```
